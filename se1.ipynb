```python
import matplotlib.pyplot as plt
import numpy as np
a = float(input("enter a values: "))
b = float(input("enter b values: "))
c = float(input("enter c values: "))
days = np.arange(1, 50, 1)
temperature_model = lambda x: a * x**2 + b * x + c
temperatures = temperature_model(days)
plt.plot(days, temperatures, label='Temperature Model')
plt.xlabel('Days')
plt.ylabel('Temperature (Â°C)')
plt.title('Quadratic Weather Modeling')
plt.legend()
plt.show()
```
